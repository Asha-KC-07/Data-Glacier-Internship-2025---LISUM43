```python
import pandas as pd
import time

# Record the start time
start_time = time.time()

try:
     # Read the CSV file
    df = pd.read_csv('Train.csv')

    # Record the end time
    end_time = time.time()
    # Calculate the loading time
    pandas_loading_time = end_time - start_time
    print(f"Pandas Loading Time: {pandas_loading_time:.4f} seconds")

    # Display few details about the DataFrame
    #print(df.head())
    print("\nDataFrame Info:")
    print(f"Number of Rows: {df.shape[0]}")
    print(f"Number of Columns: {df.shape[1]}")
    print(f"Column Names: {df.columns.tolist()}")
    print("\nDataFrame Description:")
    print(df.describe())
    print("\nDataFrame Index:")
    print(df.index)
    print("\nDataFrame Data Types:")
    print(df.dtypes)
    print("\nDataFrame Missing Values:")
    print(df.isnull().sum())
    print("\nDataFrame Duplicates:")
    print(df.duplicated().sum())
    print("\nDataFrame Memory Usage:")
    print(df.memory_usage(deep=True))
    print("\nDataFrame Sample:")
    print(df.sample(5))

    #Identify the range of date values in the first column (fecha_dato)
    # Ensure the 'fecha_dato' column is in datetime format
    df['fecha_dato'] = pd.to_datetime(df['fecha_dato'], errors='coerce')

    # Check for invalid or missing dates
    if df['fecha_dato'].isnull().any():
        print("\nWarning: Some rows have invalid or missing dates in 'fecha_dato'.")

    # Find the range of date values
    min_date = df['fecha_dato'].min()
    max_date = df['fecha_dato'].max()

    print(f"\nDate Range in 'fecha_dato':")
    print(f"Start Date: {min_date}")
    print(f"End Date: {max_date}")
    
except FileNotFoundError:
    print("Error: 'Train.csv' not found. Please ensure the file exists in the current directory.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
